```python
def find_unique_characters(data):
    unique_chars = set()
    for item in data:
        for key, value in item.items():
            unique_chars.update(value)
    return unique_chars
```

# Transliterating Devnagri into English

```python
def transliterate_to_latin(word):
    iast_mapping = {
        'á': 'aa',
        'ñ': 'n',
        'ā': 'aa',
        'ă': 'aa',
        'ī': 'i',
        'ĭ': 'i',
        'Ś': 'sh',
        'ś': 'sh',
        'Ū': 'u',
        'ū': 'u',
        'ḍ': 'd',
        'Ḣ': '',
        'ḥ': '',
        'ṃ': 'm',
        'ṅ': 'n',
        'ṇ': 'n',
        'ṛ': 'ri',
        'ṝ': 'ri',
        'ṣ': 'sh',
        'ṭ': 't'
    }

    latin_word = ''
    for char in word:
        if char in mapping:
            latin_word += iast_mapping[char]
        else:
            latin_word += char
    return latin_word



# Words to be transliterated
words = [
    'Dhṛtarāṣṭra II',
    'Duḥśāsana2',
    'Śakuni',
    'Aśvatthāman',
    'Viṣṇu',
    'śiṣyau'
]

# Transliterate each word
for word in words:
    latin_word = transliterate_to_latin(word, iast_mapping)
    print(f"{word} -> {latin_word}")
```
